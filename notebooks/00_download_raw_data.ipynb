```python
from entsoe import EntsoePandasClient
import pandas as pd
import os

API_KEY = "0b565493-fc2d-4b05-a7aa-d272f4d5251a"
client = EntsoePandasClient(api_key=API_KEY)
os.makedirs("../data/raw", exist_ok=True)
```

```python
years = [2018, 2019, 2020, 2021, 2022, 2023]

for year in years:
    start = pd.Timestamp(f"{year}-01-01", tz="Europe/Brussels")
    end   = pd.Timestamp(f"{year+1}-01-01", tz="Europe/Brussels")

    print(f"Downloading {year}...")

    try:
        ts = client.query_load("PL", start=start, end=end)

        # CASE 1: ts is already a DataFrame
        if isinstance(ts, pd.DataFrame):
            df = ts.reset_index()
            df.columns = ["timestamp", "load"]

        # CASE 2: ts is a Series
        else:
            df = ts.to_frame(name="load").reset_index()
            df.columns = ["timestamp", "load"]

        df.to_csv(f"../data/raw/load_{year}.csv", index=False)
        print(f"Saved ../data/raw/load_{year}.csv")

    except Exception as e:
        print(f"Error downloading {year}: {e}")
```
